```python
count = 0
def func():
    global count
    if count == 10000: return
    print(count)
    count += 1
    func()
func()
```

Print Name N times 

```python
def func(i, n):
    if i > n: return
    print(i, "durrani")
    func(i + 1, n)
func(1, 10) # O(n) is time complexity and also space complexity is O(n)
```

print linearly from 1 to N

```python
def func(i, n):
    if i > n: return
    print(i)
    func(i + 1, n)
func(1, 10) 
```

print linearly from N to 1

```python
def func(i, n):
    if i < 1: return
    print(i)
    func(i - 1, n)
func(10, 10) 
```

print from 1 to N but use backtracking

```python
def func(i, n):
    if i < 1: return
    func(i - 1, n) # it goes to next function and next one until base condition hits and it gets back and executes next line and prints the number in opposite but quite 1 to N order
    print(i)
func(10, 10) 
```

```python
def func(i, n):
    if i > n: return
    func(i + 1, n) # it goes to next function and next one until base condition hits and it gets back and executes next line and prints the number in opposite but quite 1 to N order
    print(i)
func(1, 10)
```
