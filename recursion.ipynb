Print Name N times 

```python
def func(i, n):
    if i > n: return
    print(i, "durrani")
    func(i + 1, n)
func(1, 10) # O(n) is time complexity and also space complexity is O(n)
```

print linearly from 1 to N

```python
def func(i, n):
    if i > n: return
    print(i)
    func(i + 1, n)
func(1, 10) 
```

print linearly from N to 1

```python
def func(i, n):
    if i < 1: return
    print(i)
    func(i - 1, n)
func(10, 10) 
```

print from 1 to N but use backtracking

```python
def func(i, n):
    if i < 1: return
    func(i - 1, n) # it goes to next function and next one until base condition hits and it gets back and executes next line and prints the number in opposite but quite 1 to N order
    print(i)
func(10, 10) 
```

```python
def func(i, n):
    if i > n: return
    func(i + 1, n) # it goes to next function and next one until base condition hits and it gets back and executes next line and prints the number in opposite but quite 1 to N order
    print(i)
func(1, 10)
```

parameterized way of summation of N

```python
def fun(n, sum):

    if n < 1:
        print(sum)
        return
    
    fun(n - 1, sum + n)

fun(3, 0)
```

functional way of summation of N

```python
def fun(n):
    if n < 1:
        return n

    return fun(n-1) + n

fun(2000)
```

finding factorial

```python
def fun(n):
    if n <= 1:
        return 1

    return fun(n-1) * n

fun(3)
```

two pointer method or double pointer method

```python
arr = [1,3,4,5,7,9]

for i in range(len(arr)//2):
    f = arr[i]
    l = arr[(-1-i)]
    arr[i] = l
    arr[(-1-i)] = f
print(arr)
```

```python
arr = [1,3,5,6,7]

def swap(x,y):
    arr[x], arr[y] = arr[y], arr[x]

def rev(l, r):
    print(l,r)
    if l >= r:
        return
    swap(l, r)
    rev(l + 1, r - 1)

rev(0, (len(arr) - 1))

arr
```

```python
arr = [1,3,5,6,7,3,5,6,7]

def swap(x,y):
    arr[x], arr[y] = arr[y], arr[x]

def rev(i):
    if i >= len(arr)//2:
        return
    swap(i, (len(arr) - i -1))
    
    rev(i + 1)
rev(0)
arr
```

```python
s = "aga"

def rev(i):
    if i >= len(s)//2:
        return True
    if s[i] != s[(len(s) - i -1)]:
        return False
    return rev(i + 1)

rev(0)
```

Fibonacci number

```python
def fun(n):
    if n <= 1:
        return n
    x = fun(n - 1) + fun(n - 2)
    return x

fun(3)
```

print all subsequences

```python
a = [3,2,1]
def f(i, arr: list):
    if i >= len(a):
        print(arr)
        return
    arr.append(a[i])
    f(i + 1, arr)
    arr.pop()
    f(i + 1, arr)

f(0, [])
```
