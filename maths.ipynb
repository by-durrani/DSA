Palindrome number

```python
def check_palindrome():
    input_value = int(input("Enter a Number..."))

    transformer_var = input_value

    reverse_var = 0

    while(transformer_var > 0):
        # Get last digit and add it to the number
        last_digit = (transformer_var % 10)
        reverse_var = (reverse_var * 10) + last_digit
        
        # remove last digit from number
        transformer_var = int(transformer_var / 10)

    if (reverse_var == input_value):
        return True
    return False

check_palindrome()
```

Reverse a 32 bit integer

```python
def reverse_an_integer_32_bit(x):
    if (x == 0):
        return x

    placeholder = x

    returning_val = 0
    if (placeholder < 0):
        placeholder = placeholder * -1
    
    while(placeholder > 0):
        returning_val = (returning_val * 10) + (placeholder % 10)
        placeholder = int(placeholder / 10)
    
    if returning_val < (-2**31) or returning_val > (2**31 -1):
        return 0

    if (x < 0):
        return returning_val * -1
    return returning_val

reverse_an_integer_32_bit(-122)
```

Armstrong Number

```python
import math

def armstrong_number(x):
    placeholder = x
    length = int(math.log10(x) + 1)
    result = 0
    while placeholder > 0:
        result = result  + (placeholder % 10 )**length
        placeholder = placeholder // 10
    print(result)
armstrong_number(407)
```

Print all divisions

```python
def print_all_divisions(n):
    for i in range(n + 1):
        if i == 0: continue
        if n % i == 0: print(i)
print_all_divisions(36) ## this will take O(n) time to get even good results in terms of time complexity we can identify mathematical pattern observation and get better results
```

```python
import math
def print_all_divisions(n):
    factors = []
    for i in range(1, int(math.sqrt(n) + 1)): # we can also do i * i will be like 6 * 6 <= n until that loop runs so instead of sqroot function we can also achieve this same thing 
        if n % i == 0: 
            factors.append(i)
            # this if block will make sure iterator is not equal to the factor only then it should append in list 
            if n // i != i: 
                factors.append(n // i)
    factors.sort()
    print(factors)
print_all_divisions(1024)
```

check for prime numbers

```python
import math
def print_all_divisions(n):
    count = 0
    for i in range(1, int(math.sqrt(n) + 1)): # we can also do i * i will be like 6 * 6 <= n until that loop runs so instead of sqroot function we can also achieve this same thing 
        if n % i == 0: 
            count += 1
            # this if block will make sure iterator is not equal to the factor only then it should append in list 
            if n // i != i: 
                count += 1

    if count == 2:
        print("is prime number")
    else:
        print("is not prime number")
print_all_divisions(13)
```

GCD or HCF

```python
def gcd(x,y):
    a = x
    b = y
    while a > 0 and b > 0:
        if a > b:
            a = a % b
        else:
            b = b % a
    
    if a == 0: 
        print(b)
    else:
        print(a)

gcd(5298, 954)
```
