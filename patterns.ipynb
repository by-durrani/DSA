```python
def main(n):
    start = 1
    for i in range(0, n):
        if (i % 2 == 0):
            start = 1
        else:
            start = 0
        for j in range(0, i+1):
            print(start, end="")
            start = 1 - start
        print('')
n = int(input('enter number'))
main(n)
```

```python
def main(n):
    # (n*2)-((i+1)+(i+1))
    for i in range(1, n+1):
        for j in range(0, i):
            print(j+1, end="")
        for j in range(0, (n - i) * 2):
            print(" ", end="")
        for j in range(0, i):
            print((i)-j, end="")
        print('')
n = int(input('enter number'))
main(n)
```

```python
def main(n):
    for i in range(0, n):
        for j in range(n - (i + 1)):
            print(" ", end="")
        inner_iter = 2*i + 1
        for j in range(1, inner_iter + 1):
            if j > inner_iter - i:
                print(chr(ord("A") + (inner_iter - i) - (j - i)), end='')
            else: print(chr(ord("A") + (j - 1)), end="")
        for j in range(i+1):
            print(" ", end="")
        print('')
n = int(input('enter number'))
main(n)
```

```python
def main(n):
    for i in range(0, n):
        for j in range(n - (i + 1)):
            print(" ", end="")
        char = ord("A")
        break_point = (2*i + 1) /2
        for j in range(1, (2*i + 1)+1):
            print(chr(char), end="")
            if j <= break_point:
                char += 1
            else:
                char -= 1
        for j in range(i+1):
            print(" ", end="")
        print('')
n = int(input('enter number'))
main(n)
```

```python
def main(n):
    for i in range(n):
        if i == 0 or i == n - 1:
            for j in range(n):
                print("* ", end="")
        else:
            for j in range(n):
                if j == 0 or j == n - 1:
                    print("* ", end="")
                else:
                    print("  ", end="")
        print('')
n = int(input('enter number'))
main(n)
```

```python
def main(n):
    for i in range(n):
        for j in range(n):
            if i == 0 or i == n - 1 or j == 0 or j == n - 1:
                print("* ", end="")
            else:
                print("  ", end="")
        print('')
n = int(input('enter number'))
main(n)
```

```python
def main(n):
    for i in range(2*n-1):
        for j in range(2*n-1):
            top = i
            left = j
            right = (2*n-2) - j
            bottom = (2*n-2) - i
            
            print(n - min(min(top, bottom), min(left, right)), end="")
        print('')
n = int(input('enter number'))
main(n)
```
